```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Automobile_data.csv")

# Display first 5 rows
print(df.head())
```

```python
plt.figure(figsize=(8,5))
df["price"].hist(bins=20)
plt.title("Distribution of Automobile Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["horsepower"], df["price"])
plt.title("Horsepower vs Price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.show()
```

```python
fuel_count = df["fuel-type"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(fuel_count,
        labels=fuel_count.index,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
sorted_df = df.sort_values("price")

plt.figure(figsize=(10,5))
plt.plot(sorted_df["price"])
plt.title("Price Trend")
plt.xlabel("Automobile Index")
plt.ylabel("Price")
plt.show()
```

```python
fuel = df["fuel-type"].value_counts()

plt.figure(figsize=(6,5))
fuel.plot(kind="bar")
plt.title("Number of Cars by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["horsepower"].hist(bins=15)
plt.title("Horsepower Distribution")
plt.xlabel("Horsepower")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["engine-size"], df["price"])
plt.title("Engine Size vs Price")
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.show()
```
